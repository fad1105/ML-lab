

Hierarchical clustering for iris dataset.


```python
from sklearn import datasets
iris = datasets.load_iris()
print("Features: ", iris.feature_names)
print("Labels: ", iris.target_names)
data=iris.data
```

```python
import matplotlib.pyplot as plt # library for visualization
from sklearn.decomposition import PCA # PCA library
pca = PCA(n_components=2)
pcaTr = pca.fit(data)
dataPCA = pcaTr.transform(data)
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram(linkage(dataPCA, 'single'),orientation='top',
distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Calculate silhouette_score
from sklearn.metrics import silhouette_score
ac2 = AgglomerativeClustering(n_clusters = 2,affinity='euclidean',linkage='ward')
plt.figure(figsize =(6, 6))
plt.scatter(dataPCA[:,0], dataPCA[:,1], 
           c = ac2.fit_predict(dataPCA), cmap ='rainbow')
plt.show()
print(silhouette_score(dataPCA, ac2.fit_predict(dataPCA)))
# print(ac2.fit_predict(dataPCA))
```
